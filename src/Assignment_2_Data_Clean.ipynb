**Part B:** Removing the units from the attributes.
We check for missing values and impute them based on each columns data

Handling the missing values:

**Mileage:** 2 missing values. Given the minimal missing data, we can impute with the mean.

**Engine:** 36 missing values. We can impute using the median, as engine size tends to have a normal distribution.

**Power:** 36 missing values. Imputation with the median is likely appropriate for similar reasons.

**Seats:** 38 missing values. Imputation with the mode is suitable, as seat count is often categorical.

**New_Price:** 5032 missing values, making up most of this column. Given the large proportion, we should drop this column due to lack of data.


```python
# Load dataset and import libraries
import pandas as pd
file_path = 'train.csv'
data = pd.read_csv('/content/train.csv',index_col = 0)

# Check the missing values
missing_values = data.isnull().sum()
missing_values
```

Remove units from 'Mileage', 'Engine', and 'Power' columns and keep only numerical values.

Impute missing values with suitable methods.



```python

data['Mileage'] = data['Mileage'].str.extract('(\d+\.?\d*)').astype(float)
data['Engine'] = data['Engine'].str.extract('(\d+\.?\d*)').astype(float)
data['Power'] = data['Power'].str.extract('(\d+\.?\d*)').astype(float)
data['New_Price'] = data['New_Price'].str.extract('(\d+\.?\d*)').astype(float)



data['Mileage'] = data['Mileage'].fillna(data['Mileage'].mean())
data['Engine'] = data['Engine'].fillna(data['Engine'].median())
data['Power'] = data['Power'].fillna(data['Power'].median())
data['Seats'] = data['Seats'].fillna(data['Seats'].mode()[0])
data[['Mileage', 'Engine', 'Power','Seats','New_Price']].head()
```

```python
# Drop 'New_Price' column due to excessive missing values
data.drop(columns=['New_Price'], inplace=True)

# Display the resulting data to confirm changes
data.head()
```
